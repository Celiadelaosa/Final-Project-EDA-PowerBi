### 1.1 Transformación de columnas demográficas

```python
demog_df.columns
```

De las 47 variables demográficas originales se seleccionan aquellas relevantes para el ánalisis, reduciendo dimensionalidad. Se renombran las columnas seleccionadas para mejorar la interpretabilidad.

```python
#se renombran las columnas
demog_df=demog_df.rename(columns={
    "SEQN": "person_id",
    "RIAGENDR": "gender",
    "RIDAGEYR": "age",
    "RIDRETH1": "race",
    "DMDBORN4": "country",
    "DMDEDUC2": "education_level",
    "DMDMARTL": "marital_status",
    "RIDEXPRG": "pregnancy_status",
    "DMDHHSIZ": "household_size",
    "INDHHIN2": "household_income",
    "INDFMIN2": "family_income",
    "INDFMPIR": "family_poverty_ratio"})
```

```python
#selección de las columnas relevantes
demog_df=demog_df[[
    "person_id",
    "gender",
    "age",
    "race",
    "country",
    "education_level",
    "marital_status",
    "pregnancy_status",
    "household_size",
    "household_income",
    "family_income",
    "family_poverty_ratio"]]
```

```python
#comprobamos
demog_df.head()
```

Se observa que en las columnas 'gender', 'race', 'country', 'eduaction_level', 'marital_status' y 'pregnancy_status' aparecen valores numéricos que dificultan la comprensión de los datos. 
Además, las columnas 'household_income', 'family_income' y 'poverty_ratio' contienen valores numéricos codificados.
Por ello, se aplican mapeos de código a texto.

```python
# Mapeamos códigos a texto
demog_df['gender']=demog_df['gender'].map({1: 'Male', 2: 'Female'})
demog_df['race']=demog_df['race'].map({
    1: 'Mexican American',
    2: 'Other Hispanic',
    3: 'Non-Hispanic White',
    4: 'Non-Hispanic Black',
    5: 'Other/Multi-Racial'
})
demog_df['country']=demog_df['country'].map({
    1: 'USA',
    2: 'Mexico',
    3: 'Other Spanish',
    4: 'Other non-Spanish'
})
demog_df['education_level']=demog_df['education_level'].map({
    1: 'Less than 9th grade',
    2: '9-11th grade',
    3: 'High school/GED',
    4: 'Some college/AA',
    5: 'College graduate or above'
})
demog_df['marital_status']=demog_df['marital_status'].map({
    1: 'Married',
    2: 'Widowed',
    3: 'Divorced',
    4: 'Separated',
    5: 'Never married',
    6: 'Living with partner'
})
demog_df['pregnancy_status']=demog_df['pregnancy_status'].map({1: 'Yes', 2: 'No'})
demog_df['household_income']=demog_df['household_income'].map({
    1:	' 0 to 4,999 $'	,
    2:	'5,000 to 9,999	$',
    3:	'10,000 to 14,999 $',		
    4:	'15,000 to 19,999 $',		
    5:	'20,000 to 24,999 $',		
    6:	'25,000 to 34,999 $',		
    7:	'35,000 to 44,999 $',	
	8:	'45,000 to 54,999 $',	
    9:	'55,000 to 64,999 $',		
    10:	'65,000 to 74,999 $',		
    12:	'Over 20,000 $',	
    13:	'Under 20,000 $',	
    14:	'75,000 to 99,999 $',		
	15:	'100,000 $ and Over',	
    77:	'Refused',		
	99:	'Dont know'
})

```

```python
#Comprobamos
demog_df.head()
```

```python
demog_df.info()
```

### 1.2 Tratamiento de valores nulos

Atendiendo al output del código anterior, se observa que no todas las columnas contienen valores nulos. Para las columnas categóricas, se realiza una imputación con 'unknown'; para las columnas numéricas, al no haber un número muy elevado de nulos, se imputa con la mediana. 

```python
#imputamos con unknown en las columnas categóricas
categoric_column=['country', 'education_level', 'marital_status', 'pregnancy_status','household_income']
demog_df[categoric_column] = demog_df[categoric_column].fillna('Unknown')

#imputamos con la mediana en las columnas numéricas
#para no ir una por una creamos una lista
numeric_column=['household_size','family_income','family_poverty_ratio']
demog_df[numeric_column] = demog_df[numeric_column].fillna(demog_df[numeric_column].median())
```

```python
#comprobamos que no hay valores nulos
demog_df.isnull().sum()
```

### 1.3 Comprobación de los valores únicos

```python
demog_df['person_id'].nunique()
```

## 4. Limpieza del dataset question

```python
#visualización general de los datos
question_df.head()
```

### 4.1 Transformación de las columnas de cuestionario

```python
#obtenemos las columnas
question_df.columns
```

Se seleccionan y renombran las columnas más relevantes.

```python
#se renombran las columnas
question_df=question_df.rename(columns={
    "SEQN": "person_id",
    "PAQ620": "moderate_activity_day",
    "PAQ605": "vigorous_activity_day",
    "ALQ120U": "alcohol_frequency",
    "SLD010H": "sleep_hours",
    "WHQ030M": "self_reported_weight",
    "WHQ500": "trying_to_do_abt_weight"
      })
```

```python
question_df=question_df[[
    "person_id",
    "moderate_activity_day",
    "vigorous_activity_day",
    "alcohol_frequency",
    "sleep_hours",
    "self_reported_weight",
    "trying_to_do_abt_weight"
]]
```

```python
question_df.head()
```

Al igual que con el primer datset se realiza un mapeo de los datos para interpretar adecuadamente los valores.

```python
# Mapeamos códigos a texto
question_df['moderate_activity_day']=question_df['moderate_activity_day'].map({
    1: 'Yes',
    2: 'No',
    7: 'Refused',
    9: 'Dont know'
})
question_df['vigorous_activity_day']=question_df['vigorous_activity_day'].map({
    1: 'Yes',
    2: 'No',
    7: 'Refused',
    9: 'Dont know'
})
question_df["self_reported_weight"]=question_df["self_reported_weight"].map({
    1: 'Fat or overweight',
    2: 'Too thin',
    3: 'About the right weight',
    7: 'Refused',
    9: 'Dont know'
})
question_df["trying_to_do_abt_weight"]=question_df["trying_to_do_abt_weight"].map({
    1: 'Lose weight',
    2: 'Gain weight',
    3: 'Stay the same weight',
    4: 'Not trying to do anything about your weight',
    7: 'Refused',
    9: 'Dont know'
})
question_df["alcohol_frequency"]=question_df["alcohol_frequency"].map({
    1: 'Weekly',
    2: 'Monthly',
    3: 'Yearly',
    7: 'Refused',
    9: 'Dont know'
})
question_df["sleep_hours"] = question_df["sleep_hours"].replace(99, np.nan)
```

```python
question_df.head()
```

```python
question_df.info()
```

### 4.2 Tratamiento de valores nulos

Para el tratamiento de valores nulos, se realiza una imputación con 'unknown' en las columnas categóricas y con la mediana para las columnas numéricas. 

```python
#imputamos con unknown en las columnas categóricas
quest_categoric_column=["moderate_activity_day",
    "vigorous_activity_day",
    "alcohol_frequency",
    "self_reported_weight",
    "trying_to_do_abt_weight"]
question_df[quest_categoric_column] = question_df[quest_categoric_column].fillna('Unknown')

#imputamos con la mediana en las columnas numéricas
question_df["sleep_hours"] = question_df['sleep_hours'].fillna(question_df['sleep_hours'].median())
```

```python
# y comprobamos
question_df.info()
```

```python
question_df['person_id'].nunique()
```

## 5. Unión de los datasets

Una vez terminada la fase de carga y limpieza, se realiza la unión de los cuatro datasets. Para ello, se emplean uniones de tipo inner join lo que garantiza que el dataset final incluya únicamente participantes con información completa en todos los conjuntos de datos, facilitando un análisis consistente.

```python
nutrition_df= demog_df.merge(diet_df, on='person_id', how='inner')
nutrition_df= nutrition_df.merge(exam_df, on='person_id', how='inner')
nutrition_df= nutrition_df.merge(question_df, on='person_id', how='inner')
```
